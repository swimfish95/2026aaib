```cpp
//week05-2a
#include <stdio.h>
int main()
{
         int a,b,c;
         scanf("%d",&a);
         scanf("%d",&b);
         scanf("%d",&c);
         int big=a,small=a;
         if (b>big) big=b;
         if (c>big) big=c;

         if (b<small) small=b;
         if (c<small) small=c;
         printf("%d\n",big-small);
}
```

```cpp
//week05-2b
#include <stdio.h>
int main()
{
    int a[3];
    int big= -99999999,small=99999999;
    for (int i =0;i<3;i++){
        scanf("%d",&a[i]);
        if (a[i]>big)big=a[i];
        if (a[i]<small)small=a[i];
    }
    printf("%d\n",big-small);
}
```

```cpp
//week05-3a
#include <stdio.h>
int main()
{
    int a[10],N=10;
    for (int i=0;i<10;i++){
        scanf("%d",&a[i]);
        if (a[i]==0){
           N=i;
           break;
        }
    }
    for(int i=N-1;i>=0;i--){
        printf("%d ",a[i]);
    }
    printf("\n");
}
```

```cpp
//week05-3b
#include <stdio.h>
int main()
{
    int a[11],N=10;
    for (int i=0;i<11;i++){
        scanf("%d",&a[i]);
        if (a[i] ==0){
           N=i;
           break;
        }

    }
    int now;
    scanf("%d",&now);
    int ans=0;
    for (int i=0;i<N;i++){
        if (a[i]==now)ans++;
    }
    printf("%d",ans);
    printf("\n");
}
```

```java
// week05-4 好玩的程式設計 for迴圈、陣列
int [] a = {10, 20, 30, 40}; // Java 的陣列

size(401, 101); // 視窗大小

for (int i=0; i<4; i++) {
    fill(#FFFFF2); // 淡黃色
    rect( i*100, 0, 100, 100);
    
    fill(0); // 黑色
    textSize(30);
    text( "a[i]:" + a[i], i*100, 50);
}
```

```java
// week05-5 好玩的程式設計
int [] a = {10, 20, 30, 40, 50, 60, 70};
int [] b = {0, 0, 0, 0, 0, 0, 0};

void setup() {
  size(700, 100);
}

void draw() {
  for (int i=0; i<7; i++) {
    if (b[i]==0) {
      fill(#FFFFF2); // 淡黃色
      rect( i*100, 0, 100, 100);
    } else {
      fill(#F5CBFF); // 粉紫色
      rect( i*100, 0, 100, 100);
    }
  }
}

void mousePressed() {
  int i = mouseX/100;
  if (b[i]==0) b[i] = 1;
  else b[i] = 0;
}
```
